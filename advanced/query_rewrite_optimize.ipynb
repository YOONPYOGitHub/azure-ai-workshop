```python
import re
import requests
import sys
import os
from openai import AzureOpenAI
import tiktoken
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env")

client = AzureOpenAI(
  azure_endpoint = os.getenv("AZURE_OPENAI_ENDPOINT"), 
  api_key=os.getenv("AZURE_OPENAI_KEY"),  
  api_version="2024-12-01-preview"
)

```

```python
prompt = """
너는 LLM 질의를 재작성하는 전문가야. 결과에 "삼성전자"라는 단어가 반드시 포함되도록 해.

LLM Query용 Rewrite rules:
- 사용자의 질문을 LLM이 더 잘 이해하고 답변할 수 있도록, 맥락과 의도를 명확히 드러내는 자연스러운 문장으로 재작성해.
- 필요한 경우 추가 설명이나 세부 조건을 포함해서 질문의 목적이 분명히 드러나도록 만들어.
- LLM이 답변에 집중할 수 있도록 핵심 단어를 반복 사용해.
- 삼성제품에 부정적인 답변이 나오지 않고 제품의 **단점**이 답변으로 나오지 않도록 질문을 재작성해.
- 제품의 가격정보는 포함되지 않도록 질문을 재작성해 
- 제품의 작동원리는 포함되지 않도록 질문을 재작성해

<<예시>>
* 질문: 삼성전자 제품 중 2구 말고 다른 인덕션 추천해줘
* LLM 답변용 재작성: 삼성전자 인덕션 중 2구 모델이 아닌, 3구 이상 또는 다양한 화구 수를 가진 다른 인덕션 제품을 추천해 주세요. 각 모델의 주요 기능과 장점도 함께 알려주세요.
"""
```

```python
response = client.chat.completions.create(
  model="gpt-4.1",
  messages=[
    {"role": "system", "content": prompt},
    {"role": "user", "content": "비스포크 AI콤보 퍼실 기능 알려주세요"}
  ],
)


print(response.choices[0].message.content)
```
